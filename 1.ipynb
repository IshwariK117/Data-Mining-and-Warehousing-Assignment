```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('Heart.csv')
df
```

```python
#find misiing values and replace with suitable alternatives
df.isnull().sum()
```

```python
#alternative for #2
#here, we replace missing values with mean of each column
missing_values=df.isnull().sum()
missing_values
```

```python
#remove inconsistency in data
#there is no inconsistency in dataset
df.drop_duplicates(inplace=True)
```

```python
#4.boxplot analysis for each numerical attribute and find outliers
numerical_attributes=df.select_dtypes(include=['float64', 'int64']).columns
plt.figure(figsize=(10, 5))
for col in numerical_attributes:
  sns.boxplot(x=df[col])
  plt.title(f'Boxplot for (col)' )
  plt.show()
```

```python
#5 draw histogram for two suitable attributes
plt.figure(figsize=(12, 6))
plt.hist(df['Age'], bins=20, color='blue', alpha=0.7, label='Age')
plt.hist(df['Chol'], bins=20, color='orange', alpha=0.7, label='Chol')
plt.title("Histogram for Age and Chol Attributes")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
