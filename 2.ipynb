```python
import pandas as pd
df=pd.read_csv('Heart.csv')
print(df.to_string()) #is a method call that converts the DataFrame to a string representation.
```

```python
#1.FIND THE STANDARD DEVIATION,VARINACE OF EVERY NUMERICAL ATTRIBUTE
#to convert all columns of the DataFrame df to numeric type
df=df.apply(pd.to_numeric,errors='coerce')
print(df.std())   #standard deviation
```

```python
print(df.var())  #variance
```

```python
#2.find covarience and perform correlation analysis using correlation coefficient
print(df.cov())  #covariance
```

```python
print(df.corr())  #correlation
```

```python
#3.how many independent features are present on given dataset
features=df.drop(columns=['AHD'])
#get the number of indepnedent features
independent_features=features.shape[1]   #features.shape:retrive he number of column ,corresponds to independent features
print(independent_features)
```

```python
#4.can we identify unwanted features
#refer to columns (or attributes) in your dataset that do not contribute meaningful information or value to the task
#features are typically either redundant, irrelevant, or provide no variability
import pandas as pd
constant_cols=[col for col in df.columns if df[col].nunique()==1]
#identify duplicate columns
duplicate_cols=[]
for i in range(len(df.columns)):
    col1=df.columns[i]
    for col2 in df.columns[i+1:]:
        if df[col1].equals(df[col2]):
            duplicate_cols.append(col2)

#combine all unwanted columns
unwanted_cols=set(constant_cols+duplicate_cols)
print(unwanted_cols)
```

```python
#5.perform the data discretization using equi frequency binning method on age attribute
bins=pd.cut(df['Age'],bins=10,labels=False)   #function to discretize the 'Age' column of your DataFrame into 10 bins (bins=10).
df['Age']=bins
print(df.head(5))
```

```python
#6.Normalize RestBP,chol,and MaxHR attribute using min-max,z-score,decimal scaling
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
#select attribute to normalize
attribute_to_normalize=['RestBP','Chol','MaxHR']

#min-max normalization
min_max_scaler=MinMaxScaler()
ans=df[attribute_to_normalize]=min_max_scaler.fit_transform(df[attribute_to_normalize])
print(ans)
#
```

```python
#z-score normalization
z_score_scaler=StandardScaler()
ans=df[attribute_to_normalize]=z_score_scaler.fit_transform(df[attribute_to_normalize])
print(ans)
#
```

```python
#decimal scaling
decimal_scaler=RobustScaler()
ans=df[attribute_to_normalize]=decimal_scaler.fit_transform(df[attribute_to_normalize])
print(ans)
```
